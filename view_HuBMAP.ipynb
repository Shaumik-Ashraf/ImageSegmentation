```python
from PIL import Image
from matplotlib import pyplot as plt
Image.MAX_IMAGE_PIXELS = 900815608
im = Image.open('0486052bb.tiff')
```

```python
arrIm = np.array(im)
arrIm.shape
```

```python
polys = []
for index in range(j2.__len__()):
    geom = np.array(j2[index]['geometry']['coordinates'])
    polys.append(geom)
```

```python
mask_1 = np.zeros(arrIm.shape)
for i in range(len(polys)):
    cv2.fillPoly(mask_1, polys[i], i+1)
```
